```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Zomato data .csv')
print(data)
```

```python
data['rate'] = data['rate'].str.split('/').str.get(0)
```

Type of restaurants

```python
sns.countplot(x = data['listed_in(type)'])
plt.xlabel('type of restaurant')
```

Conclusion:-majority of the restaurant falls in dinning category

```python
groupdata = data.groupby(['listed_in(type)'])['votes'].sum()
abdul = pd.DataFrame({'votes':groupdata})
plt.plot(abdul,marker = 'o',c = 'g')
plt.xlabel('type of restaurant',c = 'm')
plt.ylabel('Votes',c = 'm')
plt.show()
```

Conclusion:- Dinning restaurant has received maximum votes

```python
plt.hist(data['rate'],bins = 5)
plt.title('Rating Distribution')
plt.show()
```

Average order spending by couple

```python
couple_data = data['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

Conclusion:- The majority of couple preferr restaurant with an approximate of 300 rupees

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data= data)
```

```python
pivot_table = data.pivot_table(index = 'listed_in(type)',columns = 'online_order',aggfunc = 'size',fill_value = 0)
sns.heatmap(pivot_table,annot = True,cmap = 'YlGnBu',fmt = 'd')
plt.title('Heatmap')
plt.xlabel('Online_Order')
plt.ylabel('lished_in(type)')
plt.show()
```

CONCLUSION:- Dinning restaurantprimarily accept off orders whare as cafes primarily receive online this suggests that client preferr order in persin as restaurants, but preferr online ordering at cafes
